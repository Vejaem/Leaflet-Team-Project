### Creating starburst chart for NCAA season 2019

```python
# Dependencies
import pandas as pd
```

```python
#read in csv file, utitilizing read.csv method
cbb_df = pd.read_csv("cbb.csv")
cbb_df
```

```python
#filtering down to one year of data, using dot notation and a conditional statement "2019"
cbb_df = cbb_df[cbb_df.YEAR == 2019]
cbb_df
```

```python
#filtering down to the 3 columns we need .loc uses a list [rows,columns]
cbb_df = cbb_df.loc[ : , ["TEAM", "CONF", "POSTSEASON"]]
cbb_df.info()
```

```python
cbb_df.head(100)
```

```python
#remove postseason NAN using isna() method and a conditional statement
cbb_df.loc[cbb_df["POSTSEASON"].isna(), "POSTSEASON"] = ""
cbb_df.head(70)
```

```python
# what are the unique conferences, using unique and tolist methods
unique_conf = cbb_df.CONF.unique().tolist()
unique_conf
```

```python
#how many unique conferences are there?
len(unique_conf)
```

```python
#create new columns called ids, labels, parents
cbb_df["parents"] = cbb_df.CONF + "-" + "NCAA" 
cbb_df["ids"] = cbb_df.parents + "-" + cbb_df.TEAM.str.strip() 
cbb_df["labels"] = cbb_df.TEAM.str.strip()
cbb_df
```

```python
#add string strip method to NCAA
cbb_df.parents.str.strip()
cbb_df
```

```python
#creating postseason dataframe
postseason_df = cbb_df.copy()
postseason_df
```

```python
#adding 3 new columns for postseason play
postseason_df["postseason_parents"] = cbb_df.ids.str.strip()
postseason_df["postseason_ids"] = cbb_df.parents + "-" + cbb_df.POSTSEASON.str.strip()
postseason_df["postseason_labels"] = cbb_df.TEAM + "-" + cbb_df.POSTSEASON.str.strip()

postseason_df.sort_values(by="POSTSEASON", ascending=False, inplace=True)

postseason_df.reset_index(inplace=True)

postseason_df
```

```python
#remove postseason labels for teams that didnt play postseason ball (rows 68-353)
index = postseason_df[postseason_df["POSTSEASON"] == ""].index

postseason_df.loc[index, "postseason_parents"] = ""
postseason_df.loc[index, "postseason_ids"] = ""
postseason_df.loc[index, "postseason_labels"] = ""


postseason_df.head(70)
```

```python
#grabbing and relabeling the columns we need
just_postseason_df = postseason_df[["postseason_ids", "postseason_labels", "postseason_parents"]]

just_postseason_df = just_postseason_df.loc[:67, :]

just_postseason_df.rename(columns={"postseason_parents": "parents", "postseason_ids": "ids", "postseason_labels": "labels"}, inplace=True)

just_postseason_df
```

```python
#create a df for the NCAA circle
d = {'ids':["NCAA"], 'labels':["NCAA"], 'parents':[""]}
df_0 = pd.DataFrame(data=d)
df_0
```

```python
#remove any blank spaces
df_0.ids.str.strip()
df_0.labels.str.strip()
```

```python
#create conferences dataframe
d = {'labels':unique_conf}
df_1 = pd.DataFrame(data=d)
df_1
```

```python
#add ids and parent columns to df_1
df_1["ids"] = df_1.labels + "-" + "NCAA"
df_1["parents"] = "NCAA"
df_1
```

```python
#reduce cbb_df to 3 columns

cbb_df = cbb_df.loc[ : , ["ids", "labels", "parents"]]
cbb_df
```

```python
just_postseason_df
```

```python
#combining dataframes using concat method
frames1 = [cbb_df, just_postseason_df]

cbb_sunburst_df = pd.concat(frames1)
cbb_sunburst_df
```

```python
# #add postseason df
frames2 = [df_0, df_1, cbb_sunburst_df]

cbb_sunburst_df = pd.concat(frames2)
cbb_sunburst_df
```

```python
# save df as csv using to_csv method
cbb_sunburst_df.to_csv("cbb_sunburst2019.csv", index=False)
```
