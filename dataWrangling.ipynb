```python
#Import dependencies

import pandas as pd
import json
```

```python
# import the csv data file

cbb_df = pd.read_csv("cbb.csv")
cbb_df
```

```python
# inspect the df using df.info() to look for NaN values
cbb_df.info()
```

```python
cbb_df["Perc_Wins"] = round(cbb_df["W"] / cbb_df["G"], 2)
cbb_df
```

```python
cbb_df = cbb_df.loc[:,["YEAR", "CONF", "TEAM", "W", "G", "Perc_Wins"]]
cbb_df
```

```python
# filter data by year
cbb_2019_df = cbb_df.loc[cbb_df["YEAR"] == 2019, ["CONF", "TEAM", "W", "G", "Perc_Wins"]]
                                               
cbb_2019_df
```

```python
# sort data by Perc_Wins
cbb_2019_df.sort_values(by="W", ascending=False, inplace=True)
cbb_2019_df
```

```python
# store dataframe in JSON format
cbb_2019_df.to_json("cbb_2019.json", orient = "split", compression = "infer", index = "true")
```
