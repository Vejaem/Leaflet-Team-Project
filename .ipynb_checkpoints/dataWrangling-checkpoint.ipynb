```python
import pandas as pd
import json
```

```python
# import the csv data file
cbb_df = pd.read_csv("cbb.csv")
cbb_df
```

```python
# inspect the df using df.info() to look for NaN values
cbb_df.info()
```

```python
cbb_df = cbb_df.loc[:,["YEAR", "CONF", "TEAM", "G", "W"]]
cbb_df.groupby(["YEAR", "CONF"])["G"].sum()
```

```python
# save cbb_df as a json file
result = cbb_df.to_json()
parsed = json.loads(result)
```
